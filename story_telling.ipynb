**Question 1: How income levels are related to education levels?**

```python
# Some cleaning had to be done with the categories of education
# Combined some of the categories together and renamed them
char_to_replace = {'Assoc-voc': 'Associate_degree',
                   'Assoc-acdm': 'Associate_degree',
                   'Prof-school': 'Associate_degree',
                   '11th': 'Some_HS', '9th': 'Some_HS', '5th-6th':'Some_MS',
                   '10th': 'Some_HS', '12th': 'Some_HS', '7th-8th': 'Some_MS', '1st-4th': 'ES', 'Preschool': 'ES'}
# Iterate over all key-value pairs in dictionary
for key, value in char_to_replace.items():
    # Replace key character with value character in string
    income['education'] = income['education'].replace(key, value)
```

**Question 2: How does hours-per-week and their education levels differ and also how this association differs among men and women?**

```python
plt.figure(figsize=(16,10))
ax = sns.pointplot(x="education", y="hours-per-week", hue="gender",
                   data=income, order=['ES','Some_MS','Some_HS','HS-grad','Some-college','Associate_degree','Bachelors','Masters', 'Doctorate'])
plt.title("Count of individuals in different workclass distributed based on their gender")
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.show()
```

Ans: The above pointplot shows that on average, males work more than females during a week and this difference of hours doesn't change much at different levels of education. 
Interestingly, as education level increases so does the hours of work/week. Another thing is for those with masters and doctorate the gap in hours of work between men and women becomes smaller.

**Question 4: How hours-per-week vary among different workclasse?**

```python
plt.figure(figsize=(16,10))
plt.title("Hours-of-week and workclass distribution")
sns.violinplot(data=income, x="workclass", y="hours-per-week")
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.show()

```

median number of hours that people work is high among the self-employed-incorporated group, also in this group individuals differ much in their number of hours, as evident from the shallow distribution. Those who are in the private, local government, state government and federal government show a narrow distribution meaning majority of individuals' working hours do not differ from each other. This could be true because in this work classes they need to follow strict work hours, unlike in self-employed groups where hours of work is more flexible.

**Question 5: How income is dispersed among men and women?**

```python
plt.figure(figsize=(16,10))
gender_counts = (income.groupby(['income'])['gender']
                     .value_counts(normalize=True)
                     .rename('percentage')
                     .mul(100)
                     .reset_index())
p = sns.barplot(x="gender", y="percentage", hue="income", data=gender_counts)
plt.title('Income and Gender')
_ = plt.setp(p.get_xticklabels(), rotation=45)  
```

Among those who make more than 50K, 80% are men and 20% are women. Among those who make less than 50k, men make around 60% and women make around 40%. 
In general the income disparity due to gender is higher among higher income people.

**Question 6: How hours of work and age differ?, are there any other variations in this association due to gender**

```python
plt.figure(figsize=(16,10))
plt.title("Hours-of-week, Age and Gender Scatterplot")
sns.scatterplot(data=income, x="age", y="hours-per-week", hue='gender')
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.show()
```


During age 17 to 25 or so, the work hours differ hugely as most of them would be in college as students or as only part time workers. Later when they reach the age of typical work force, their work hours increases, and there's a steady maintenance of their working hours in the range of 40 to 60(mostly). Later when they age, the number of hours falls. This, we can find from the empty white space (without dots) on the upper right hand corner. However, few people in the 90s age range mentioned they work for higher number of hours!
